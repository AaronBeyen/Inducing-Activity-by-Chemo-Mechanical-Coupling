```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
```

```python
import pandas as pd

# Read data
Bw0 = pd.read_csv('Bw0.csv', header=None, names=['i', 'berekening'])
Bw1 = pd.read_csv('Bw1.csv', header=None, names=['i', 'berekening'])
Bw3 = pd.read_csv('Bw3.csv', header=None, names=['i', 'berekening'])
Bw5 = pd.read_csv('Bw5.csv', header=None, names=['i', 'berekening'])
Bw8 = pd.read_csv('Bw8.csv', header=None, names=['i', 'berekening'])

vw0 = pd.read_csv('vw0.csv', header=None, names=['i', 'berekening'])
vw1 = pd.read_csv('vw1.csv', header=None, names=['i', 'berekening'])
vw3 = pd.read_csv('vw3.csv', header=None, names=['i', 'berekening'])
vw5 = pd.read_csv('vw5.csv', header=None, names=['i', 'berekening'])
vw8 = pd.read_csv('vw8.csv', header=None, names=['i', 'berekening'])

# Convert to NumPy array
Bw0_array = Bw0.to_numpy()
Bw1_array = Bw1.to_numpy()
Bw3_array = Bw3.to_numpy()
Bw5_array = Bw5.to_numpy()
Bw8_array = Bw8.to_numpy()

vw0_array = vw0.to_numpy()
vw1_array = vw1.to_numpy()
vw3_array = vw3.to_numpy()
vw5_array = vw5.to_numpy()
vw8_array = vw8.to_numpy()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
plt.plot(Bw0_array[:, 0], Bw0_array[:, 1], label = r'$\beta w = 0$', color = 'blue')
plt.plot(Bw1_array[:, 0], Bw1_array[:, 1], label = r'$\beta w = 1$', color = 'red')
plt.plot(Bw3_array[:, 0], Bw3_array[:, 1], label = r'$\beta w = 3$', color = 'orange')
plt.plot(Bw5_array[:, 0], Bw5_array[:, 1], label = r'$\beta w = 5$', color = 'lime')
plt.plot(Bw8_array[:, 0], Bw8_array[:, 1], label = r'$\beta w = 8$', color = 'cyan')
ax.set_ylabel(r"$\frac{B(x)}{B_c}$", fontsize = 35, rotation = 0,labelpad=35)
ax.set_xlabel(r"$x/L$", fontsize = 30)
#ax.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=0.7)
ax.legend(fontsize = 20, loc = 'upper right')
plt.tight_layout()
ax.spines[['right', 'top']].set_visible(False)
ax.tick_params(axis='both', which='major', labelsize=15)
for axis in 'left', 'bottom':
    ax.spines[axis].set_linewidth(1.5)
# Move x-axis to y = 0
ax.spines['bottom'].set_position(('data', 0))

# Ensure ticks are on the moved axis
ax.xaxis.set_ticks_position('bottom')
#plt.savefig('Bw.png')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
plt.plot(vw0_array[:, 0], vw0_array[:, 1], label = r'$\beta w = 0$', color = 'blue')
plt.plot(vw1_array[:, 0], vw1_array[:, 1], label = r'$\beta w = 1$', color = 'red')
plt.plot(vw3_array[:, 0], vw3_array[:, 1], label = r'$\beta w = 3$', color = 'orange')
plt.plot(vw5_array[:, 0], vw5_array[:, 1], label = r'$\beta w = 5$', color = 'lime')
plt.plot(vw8_array[:, 0], vw8_array[:, 1], label = r'$\beta w = 8$', color = 'cyan')
ax.set_ylabel(r"$\frac{\nu(x)}{\nu_c}$", fontsize = 35, rotation = 0,labelpad=35)
ax.set_xlabel(r"$x/L$", fontsize = 30)
#ax.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=0.7)
ax.legend(fontsize = 20, loc = 'upper right')
plt.tight_layout()
ax.spines[['right', 'top']].set_visible(False)
ax.tick_params(axis='both', which='major', labelsize=15)
for axis in 'left', 'bottom':
    ax.spines[axis].set_linewidth(1.5)
# Move x-axis to y = 0
ax.spines['bottom'].set_position(('data', 0))
# Ensure ticks are on the moved axis
ax.xaxis.set_ticks_position('bottom')
#plt.savefig('vw.png')
plt.show()
```
